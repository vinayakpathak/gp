```python
import pandas as pd
import numpy as np
```

```python
curve_code = """
data {
    int<lower=0> N;
    real x[N];
    vector[N] mu;
    real scale;
    real v;
}
transformed data {
  matrix[N, N] K;
  for (i in 1:(N - 1)) {
    K[i, i] = (1 + 0.1)*v;
    for (j in (i + 1):N) {
      K[i, j] = v*exp(-0.5 / scale * square(x[i] - x[j]));
      K[j, i] = K[i, j];
    }
  }
  K[N, N] = (1 + 0.1)*v;
}
parameters {
  vector[N] y;
}
model {
  y ~ multi_normal(mu, K);
}
"""


sm = pystan.StanModel(model_code=curve_code)
```

```python
curve_dat = {'N': 30,
               'x': range(1, 31),
             'mu': [0] * 30,
             'scale': 30,
             'v': 0.1
              }

fit = sm.sampling(data=curve_dat, iter=1000, chains=4)
```

```python
la = fit.extract(permuted=True)  # return a dictionary of arrays
#mu = la['mu']

## return an array of three dimensions: iterations, chains, parameters
a = fit.extract(permuted=False)

la
```

```python
def plot_curves(xs, curves, ymin, ymax):
    for curve in curves:
        plt.plot(xs, curve, 'k', alpha=0.1)
    plt.ylim(ymin, ymax)
    plt.show()
```

```python
plot_curves(curve_dat['x'], la['y'][1:300], ymin=-5, ymax=5)
```

```python
curve_pca_code = """
data {
  int<lower=0> D;  //number of factors
  int<lower=0> N;  //original dimensions
  int<lower=0> M;  //number of observations
  real sigma;
  
}
parameters {
  matrix[N, D] W;
  vector[N] mu;
  vector[D] z;
}
model {
  z ~ multi_normal(rep_vector(0, D), diag_matrix(rep_vector(1, D)));
  y ~ multi_normal(W*z+mu, diag_matrix(rep_vector(sigma*sigma, N)));
  y_obs ~ normal((W*z+mu)[i], sigma*sigma);
}
"""

pcasm = pystan.StanModel(model_code=curve_pca_code)
```
